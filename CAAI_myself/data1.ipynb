```python
user_eventft = all_train.merge(user,how="left",left_on="inviter_id",right_on="user_id")
```

```python
user_eventft = user_eventft.merge(user,how="left",left_on="voter_id",right_on="user_id",suffixes=["_inviter","_voter"])
```

```python
user_eventft =user_eventft.drop(["user_id_inviter","user_id_voter"],axis=1)
```

```python
user_eventft
```

```python
user_eventft["month"] = user_eventft["ds"]/100
user_eventft["month"] = user_eventft["month"].astype("int")
```

```python
user_eventft["month"].value_counts()
```

```python
a = user_eventft.groupby(["event_id","age_level_inviter"])["age_level_inviter"].count().unstack().reset_index()
b = user_eventft.groupby(["event_id","user_level_inviter"])["user_level_inviter"].count().unstack().reset_index()
c = user_eventft.groupby(["event_id","user_level_voter"])["user_level_voter"].count().unstack().reset_index()
d = user_eventft.groupby(["event_id","age_level_voter"])["age_level_voter"].count().unstack().reset_index()
e = user_eventft.groupby(["event_id","month"])["month"].count().unstack().reset_index()
eventft = eventft.merge(a,on="event_id",how="left")
eventft = eventft.merge(b,on="event_id",how="left")
eventft = eventft.merge(c,on="event_id",how="left")
eventft = eventft.merge(d,on="event_id",how="left")
eventft = eventft.merge(e,on="event_id",how="left")
```

```python
a = user_eventft[user_eventft["if_inviter_participate"]==1].groupby(["event_id","age_level_inviter"])["age_level_inviter"].count().unstack().reset_index()
b = user_eventft[user_eventft["if_inviter_participate"]==1].groupby(["event_id","user_level_inviter"])["user_level_inviter"].count().unstack().reset_index()
c = user_eventft[user_eventft["if_inviter_participate"]==1].groupby(["event_id","user_level_voter"])["user_level_voter"].count().unstack().reset_index()
d = user_eventft[user_eventft["if_inviter_participate"]==1].groupby(["event_id","age_level_voter"])["age_level_voter"].count().unstack().reset_index()
e = user_eventft[user_eventft["if_inviter_participate"]==1].groupby(["event_id","month"])["month"].count().unstack().reset_index()
eventft = eventft.merge(a,on="event_id",how="left")
eventft = eventft.merge(b,on="event_id",how="left")
eventft = eventft.merge(c,on="event_id",how="left")
eventft = eventft.merge(d,on="event_id",how="left")
eventft = eventft.merge(e,on="event_id",how="left")
```

```python
a = user_eventft[user_eventft["if_voter_participate"]==1].groupby(["event_id","age_level_inviter"])["age_level_inviter"].count().unstack().reset_index()
b = user_eventft[user_eventft["if_voter_participate"]==1].groupby(["event_id","user_level_inviter"])["user_level_inviter"].count().unstack().reset_index()
c = user_eventft[user_eventft["if_voter_participate"]==1].groupby(["event_id","user_level_voter"])["user_level_voter"].count().unstack().reset_index()
d = user_eventft[user_eventft["if_voter_participate"]==1].groupby(["event_id","age_level_voter"])["age_level_voter"].count().unstack().reset_index()
e = user_eventft[user_eventft["if_voter_participate"]==1].groupby(["event_id","month"])["month"].count().unstack().reset_index()
eventft = eventft.merge(a,on="event_id",how="left")
eventft = eventft.merge(b,on="event_id",how="left")
eventft = eventft.merge(c,on="event_id",how="left")
eventft = eventft.merge(d,on="event_id",how="left")
eventft = eventft.merge(e,on="event_id",how="left")
```

```python
eventft = eventft.fillna(0)
```
